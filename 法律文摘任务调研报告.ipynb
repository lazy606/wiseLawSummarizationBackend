## 一、任务
从翻译成中文的裁判文书得到案情摘要信息（以此为例），并将信息组合生成相应字段内容。









## 二、解决方法










### 1.NLP中的阅读理解任务（MRC）
· 内容：作为NLP的篇章抽取型阅读理解（Span-Extraction）（[出处](https://github.com/ymcui/Chinese-ELECTRA)）任务处理，考虑到可能涉及到案情摘要信息可能涉及裁判文书中的多处，可以作为多答案篇章抽取型阅读理解(Multi-Span Extraction)([出处](https://zhuanlan.zhihu.com/p/101248202))任务处理。
· 参考文章：[Span-Extraction](https://paperswithcode.com/task/span-extraction-mrc)
        [Multi-Span Extraction](https://paperswithcode.com/search?q_meta=&q_type=&q=Multi-Span+Extraction)
· 可用数据集:法律阅读理解数据集[CAIL](https://github.com/china-ai-law-challenge/CAIL2019/tree/master/%E9%98%85%E8%AF%BB%E7%90%86%E8%A7%A3)、通用阅读理解数据集[cmrc2018](https://github.com/ymcui/cmrc2018)
· 可用模型:中文通用模型[Chinese-ELECTRA](https://github.com/ymcui/Chinese-ELECTRA)(非Multi-Span Extraction)






### 2.采用wordembedding取代正则表达式做内容抽取再扩写
· 内容：使用词嵌入模型将fact等词（法学院提供）通过词嵌入模型转为向量，再将裁判文书切分成句子转成向量，然后将选用词与裁判文书句子的向量取余弦相似度，找到一个阈值去掉完全无关句子，然后将得到的相关句子做一个扩写。
· 法律词嵌入模型：legal-ELECTRA(出处)[https://github.com/ymcui/Chinese-ELECTRA] ， Lawformer(出处)[https://github.com/thunlp/LegalPLMs]
